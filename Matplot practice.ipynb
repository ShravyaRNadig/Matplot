```python
x=[10,20,30,40,50]
y=[22,24,26,28,30]
z=[35,45,50,60,70]
```

```python
plt.plot(x,y,"o")
plt.plot(x,z,"^")
```

```python
plt.plot(x,y,"o")
plt.plot(x,z,"^")
plt.show()
```

```python
plt.plot(x,y,"o")
plt.show()
plt.plot(x,z,"^")
plt.show()
```

```python
plt.plot(x,y,"o")
plt.title("line graph")#Giving tirle
plt.xlabel("Age")#x value (axis)
plt.ylabel("Height")#y value(axis)
```

```python
plt.plot(x,y,"r")#giving color r-red
plt.title("line graph")#Giving title n convays info being shown in graph
plt.xlabel("Age")#x value (axis)
plt.ylabel("Height")#y value(axis)
```

```python
plt.plot(x,y,"r")#giving color r-red
plt.title("line graph")
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()
```

```python
plt.plot(x,y,"g")#giving color g-green
plt.title("line graph")
plt.xlabel("Age")
plt.ylabel("Height")
## matlotlib automatically selects the range of axis depending on data,
##but sometimes might have to set to specific values
## can set for individual axes
## setting axis limts
plt.xlim(0,50)
plt.ylim(0,30)
plt.show()
```

```python
plt.plot(x,y,"g")#giving color g-green
plt.title("line graph")
plt.xlabel("Age")
plt.ylabel("Height")
plt.xlim(0,50)
plt.ylim(25,30)
plt.show()
```

```python
plt.plot(x,y,"orange")
plt.title("line graph")
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()
```

```python
plt.plot(x,y,"turquoise")
plt.plot(x,z,"indigo")
plt.title("line graph")
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()
```

```python
plt.plot(x,y,"brown")
plt.plot(x,z,"violet")
plt.title("line graph")
plt.xlabel("Age")
plt.ylabel("Height")
plt.text(0,40,"X vs Z")
plt.text(45,20,"X vs Y")
plt.show()
```

```python
plt.plot(x,y,"magenta")
plt.plot(x,z,"gold")
plt.title("line graph")
plt.xlabel("Age")
plt.ylabel("Height")
plt.text(10,40,"X vs Z")
plt.text(45,20,"X vs Y")
plt.show()
```
